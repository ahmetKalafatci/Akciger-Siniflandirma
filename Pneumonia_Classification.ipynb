```python
import os
import pickle
import pandas as pd
from skimage.io import imread
from skimage.transform import resize
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
input_dir = r'C:\Users\90541\Desktop\Yapay Zeka\Udemy Yapay Zeka\Machine Learning\Projeler\Akciğer Hastalık Sınıflandırması\Lung X-Ray Image'
```

```python
categories = ['Lung_Opacity', 'Normal','Viral Pneumonia']
```

```python
# imageları yükleyip piksel değerlerini alma
data = []
labels = []
for category_idx, category in enumerate(categories):
    for file in os.listdir(os.path.join(input_dir, category)):
        img_path = os.path.join(input_dir, category, file)
        img = imread(img_path)
        img = resize(img, (15, 15))
        data.append(img.flatten())
        labels.append(category_idx)

data = np.asarray(data)
labels = np.asarray(labels)
```

```python
# dataframe oluşturma
df = pd.DataFrame(data)
df['labels'] = labels
df.head()
```

```python
y = df.labels.values.reshape(-1,1)
x = df.drop(["labels"],axis = 1)
x.head()
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,  y_test = train_test_split(x,y,test_size = 0.2, random_state=42)
```

```python
# Decision Tree Modeli ile
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
print("accuracy of decisiontree algo: % ",dt.score(x_test,y_test)*100)
```

```python
# Random Forest Modeli ile
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100,random_state=53) 
rf.fit(x_train,y_train)
```

```python
print("accuracy of decisiontree algo: % ",rf.score(x_test,y_test)*100)
```
